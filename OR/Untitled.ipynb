```python
# مدل
model = Model("Single Hub Location")

# متغیرها
X = model.addVars(cities, vtype=GRB.BINARY, name="X")  # انتخاب هاب

# تابع هدف
model.setObjective(
    sum(C[i] * X[i] for i in cities) + 
    sum(C_trans * (D.get((i, j), 0) / 100) * T.get((i, j), 0) * X[i] for i in cities for j in cities),
    GRB.MINIMIZE
)

# محدودیت: فقط یک هاب باید انتخاب شود
model.addConstr(sum(X[i] for i in cities) == 1)

# حل مدل
model.optimize()


# نمایش نتایج
if model.status == GRB.OPTIMAL:
    print("هزینه کل:", model.objVal)
    
    # هاب انتخابی
    hub = None
    print("هاب انتخابی:")
    for i in cities:
        if X[i].x > 0.5:
            hub = i
            print(f" - {i}")
    
    # محاسبه هزینه شهرهای دیگر
    if hub:
        total_other_cost = 0
        print("هزینه شهرهای دیگر:")
        for j in cities:
            if j != hub:  # شهرهایی که هاب نیستند
                cost = C_trans * (D[(hub, j)] / 100) * T[(hub, j)]
                total_other_cost += cost
                print(f" - شهر {j} -> هزینه انتقال: {cost:.2f} دلار")
        
        print(f"مجموع هزینه انتقال شهرهای دیگر: {total_other_cost:.2f} دلار")
```

```python
# محاسبه هزینه کلی برای هر شهر به عنوان هاب
print("هزینه کل برای هر شهر به عنوان هاب:")
for hub in cities:
    total_cost = C[hub]  # هزینه نگهداری هاب
    transfer_cost = 0    # هزینه انتقال مسافران
    
    # محاسبه هزینه انتقال برای تمامی شهرها به هاب انتخابی
    for i in cities:
        if i != hub:  # اگر شهر i هاب نباشد
            transfer_cost += C_trans * (D[(hub, i)] / 100) * T[(hub, i)]
    
    total_cost += transfer_cost
    print(f" - اگر {hub} هاب باشد: هزینه کل = {total_cost:.2f} دلار")
```
