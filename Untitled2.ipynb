```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("EnglandWeather.csv")
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.describe()
```

```python
df.describe(include=object)
```

```python
df['Formatted Date'] = df['Formatted Date'].str.slice(0, 23)
df
```

```python
summaries=df['Summary'].drop_duplicates().to_numpy()
df2=pd.DataFrame(columns=['weather','count'])
for i in range(0,len(summaries)):
    count=df[df['Summary']==summaries[i]]['Summary'].count()
    df2=df2._append({'weather':summaries[i],'count':count},ignore_index=True)
plt.bar(df2['weather'],df2['count'])  
plt.grid()
plt.xticks(rotation=90)
plt.show()
    
```

```python
print('Rainy Days =>',df[df['Precip Type']=='rain']['Precip Type'].count())
print('Snowy Days =>',df[df['Precip Type']=='snow']['Precip Type'].count())
```

```python
plt.scatter(df['Temperature (C)'],df['Humidity']) 
plt.xlabel('Temp')
plt.ylabel('Humidity')
plt.grid()
plt.show()
```

```python
plt.scatter(df['Wind Speed (km/h)'],df['Humidity'])  
plt.xlabel('Wind Speed')
plt.ylabel('Humidity')
plt.grid()
plt.show()
```

```python
df3=df[df['Pressure (millibars)']!=0]
plt.scatter(df3['Pressure (millibars)'],df3['Humidity'])  
plt.xlabel('Pressure (millibars)')
plt.ylabel('Humidity')
plt.grid()
plt.show()
```

```python
year=pd.to_datetime(df["Formatted Date"]).dt.year
plt.xlabel('year')
sns.countplot(x=year)
```

```python
month=pd.to_datetime(df["Formatted Date"]).dt.month
plt.xlabel('month')
sns.countplot(x=month)
```

```python
day=pd.to_datetime(df["Formatted Date"]).dt.day
plt.xlabel('day')
plt.xticks(rotation=90)
sns.countplot(x=day)
```
