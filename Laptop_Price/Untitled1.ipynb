```python

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv('laptop_data.csv')
df=data.copy()
df
```

```python
df.drop(columns='Unnamed: 0',inplace=True)
```

```python
df['Weight']=df['Weight'].str[:-2]
df['Ram']=df['Ram'].str[:-2]
df.Weight=df.Weight.astype(float)
df.Ram=df.Ram.astype(float)
```

```python
df.dtypes
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df2=df.copy()
df2['Company']=le.fit_transform(df2['Company'])
df2['TypeName']=le.fit_transform(df2['TypeName'])
df2['ScreenResolution']=le.fit_transform(df2['ScreenResolution'])
df2['Cpu']=le.fit_transform(df2['Cpu'])
df2['Memory']=le.fit_transform(df2['Memory'])
df2['Gpu']=le.fit_transform(df2['Gpu'])
df2['OpSys']=le.fit_transform(df2['OpSys'])
```

```python
df2
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score

def eval_model(ytest,ypred):
    mae = mean_absolute_error(ytest,ypred)
    mse = mean_squared_error(ytest,ypred)
    rmse = mean_squared_error(ytest,ypred,squared=False)
    r2s = r2_score(ytest,ypred)
    return {'MAE':mae,'MSE':mse,'RMSE':rmse}

def model_res(model,x_train,x_test,y_train,y_test,ypred,mname):
    train_r2 = model.score(x_train,y_train)
    test_r2 = model.score(x_test,y_test)
    w = eval_model(y_test,ypred)
    res_metrics = {'Train_R2':train_r2,'Test_R2':test_r2,'Test_MSE':w['MSE'],
                    'Test_RMSE':w['RMSE'],'Test_MAE':w['MAE']}
    res = pd.DataFrame(res_metrics, index=[mname]) 
    return res, res_metrics
```

```python
x = df2.drop('Price',axis=1)
# y = df2['Price']  
y = np.log(df2['Price'])     
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=8)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr_df,lr_res = model_res(lr,x_train,x_test,y_train,y_test,lr.predict(x_test),'LinReg')
lr_df
```
