```python
df_box = df.columns
i =  0
while i < len(df_box):
    fig = plt.figure(figsize = (25, 4), facecolor='burlywood')
    
    plt.subplot (1 , 2 , 1)
    box = sns.boxplot(x=df_box [i], data=df, palette="hot_r", width=0.5, dodge=True )
    box.set_title("Boxplot of Column " + df_box [i] , fontsize=18 , backgroundcolor = 'black',color='white')
    box.set(xlabel=None)
    box.set_facecolor('white')
    plt.grid()
    i += 1

    if i < len(df_box):
        plt.subplot (1 , 2 , 2)
        box = sns.boxplot(x=df_box [i], data=df, palette="hot_r", width=0.5, dodge=True )
        box.set_title("Boxplot of Column " + df_box [i] , fontsize=18 , backgroundcolor = 'black',color='white')
        box.set(xlabel=None)
        box.set_facecolor('white')
        plt.grid()
        i += 1

    plt.show()
```

```python
x=df2.drop(columns='Anaemic')
y=df2['Anaemic'].values.reshape(-1,1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=1)
```

```python
from sklearn.linear_model import LogisticRegression
logreg=LogisticRegression()
logreg.fit(x_train,y_train)
```

```python
print('Test:',logreg.score(x_test,y_test))
print('Training:',+logreg.score(x_train,y_train))
```

```python
y_pred=logreg.predict(x_test)
```

```python
from sklearn.metrics import roc_curve
fpr,tpr,_=roc_curve(y_test,y_pred)
plt.plot(fpr,tpr)
```

```python
from sklearn.model_selection import StratifiedKFold
from yellowbrick.model_selection import CVScores
_, ax = plt.subplots()

# Create a cross-validation strategy
cv = StratifiedKFold(10)

# Create the cv score visualizer
oz = CVScores(
    LogisticRegression(), ax=ax, cv=cv, scoring='accuracy'
)
oz.fit(x_train, y_train)
oz.poof();
```

```python

# Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test,y_pred)
# confusion metric visual
import seaborn as sn
plt.figure(figsize=(7,5))
cm_df = pd.DataFrame(cm,
                     index = ['No','Yes'], 
                     columns = ['No','Yes'])
sn.heatmap(cm_df,annot=True)

plt.xlabel('Predicted',color='black',size=15)
plt.ylabel('Truth',color='black',size=15)
plt.show()
```

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
```

```python
print('Test:',nb.score(x_test,y_test))
print('Training:',+nb.score(x_train,y_train))
```

```python
y_pred=nb.predict(x_test)
```

```python
from sklearn.model_selection import StratifiedKFold
from yellowbrick.model_selection import CVScores
_, ax = plt.subplots()

# Create a cross-validation strategy
cv = StratifiedKFold(10)

# Create the cv score visualizer
oz = CVScores(
    GaussianNB(), ax=ax, cv=cv, scoring='accuracy'
)
oz.fit(x_train, y_train)
oz.poof();
```

```python

# Confusion Matrix
from sklearn.metrics import confusion_matrix,classification_report
cm = confusion_matrix(y_test,y_pred)
# confusion metric visual
import seaborn as sn
plt.figure(figsize=(7,5))
cm_df = pd.DataFrame(cm,
                     index = ['No','Yes'], 
                     columns = ['No','Yes'])
sn.heatmap(cm_df,annot=True)

plt.xlabel('Predicted',color='black',size=15)
plt.ylabel('Truth',color='black',size=15)
plt.show()
```
