```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("HAM10000_metadata.csv")
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df.describe(include=[np.object_])
```

```python
plt.scatter(df["age"],df["localization"])
plt.xlabel("Age")
plt.xticks(np.arange(0,90,5))
plt.ylabel("Cancer Location")
plt.grid()
plt.show()
```

```python
plt.scatter(df["sex"],df["localization"])
plt.xlabel("sex")
plt.ylabel("Cancer Location")
plt.grid()
plt.show()
```

```python
plt.bar(df["age"],df["localization"])
plt.xlabel("age")
plt.ylabel("Cancer Location")
plt.xticks(np.arange(0,90,5))
plt.show()
```
