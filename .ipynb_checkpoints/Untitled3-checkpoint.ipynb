```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('Market.csv')
df=pd.DataFrame(data)
df
```

```python
df=df[df['Index']=='NYA']
df=df.round()
df
```

```python
plt.figure(figsize=(25,10))
plt.xticks(np.arange(0,13948,200),size='large',rotation=90)
plt.plot(df.Date,df['Adj Close'])
plt.show()
```

```python
plt.figure(figsize=(25,10))
plt.xticks(np.arange(0,13948,200),size='large',rotation=90)
plt.plot(df.Date,df['Close'])
plt.show()
```
