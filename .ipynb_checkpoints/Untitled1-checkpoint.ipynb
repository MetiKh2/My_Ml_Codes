```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv("EnglandWeather.csv")
data
```

```python
df=pd.DataFrame(data)
df
```

```python
df['Formatted Date'] = df['Formatted Date'].str.slice(0, 23)
df
# حذف ساعت
```

```python
year=pd.to_datetime(df["Formatted Date"]).dt.year
sns.countplot(x=year)
```

```python
filtered_df = df[df['Formatted Date'].str.slice(0,4)=='2015']
filtered_df
# فیلتر کردن سال
```

```python
filtered_df=filtered_df[filtered_df['Formatted Date'].str.slice(5,7)=='06']
filtered_df
#فیلتر کردن ماه
```

```python
filtered_df['Formatted Date']=filtered_df['Formatted Date'].str.slice(8,10)
filtered_df
# جدا کردن روز های ماه 
# روز های ماه 06
```

```python
filtered_df['Formatted Date']=filtered_df['Formatted Date'].astype(int)
filtered_df.dtypes
```

```python
filtered_df.sort_values(by="Formatted Date",inplace=True)
# به ترتیب کردن روز ها
```

```python
filtered_df
```

```python
plt.scatter(filtered_df['Formatted Date'],filtered_df['Humidity'])
plt.show()
```

```python
filtered_df
```

```python
mean_df=pd.DataFrame(columns=['Day','Humidity'])
for i in range(1,31):
    mean_df=mean_df._append({'Day':i,'Humidity':np.mean(filtered_df[filtered_df['Formatted Date']==i]['Humidity'])},ignore_index=True)
mean_df
# بدست اوردن میانگین هر روز از ماه فیلتر شده
```

```python
plt.bar(mean_df['Day'],mean_df['Humidity'])
```
