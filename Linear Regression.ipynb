```python
def classify(data,col,values):
  for i,val in enumerate(values):
     data.loc[data[col] == val,[col]] =i+2
```

```python
df2=df.copy()
classify(data=df2,col='Fuel_Type',values=['Petrol', 'Diesel', 'CNG'])
classify(data=df2,col='Seller_Type',values=['Dealer', 'Individual'])
classify(data=df2,col='Transmission',values=['Automatic', 'Manual'])
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split,KFold,cross_val_score
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
```

```python
x=df2[['Age','Present_Price','Kms_Driven','Fuel_Type','Seller_Type','Transmission','Owner']]
y=df2['Selling_Price'].values.reshape(-1,1)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=0) 
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred=model.predict(x_test)
```

```python
print("Mean Squared Error",mean_squared_error(y_test,y_pred))
print("Mean Absolute Error",mean_absolute_error(y_test,y_pred))
print("R2 Score",r2_score(y_test,y_pred))
```

```python
kfold_validation=KFold(10)
results=cross_val_score(model,x,y,cv=kfold_validation)
print(results)
print(np.mean(results))
```

```python
def check(dim,test_size):
    r2=0.8793462370195175
    for col in x:
        new_col_name=col+str(dim)
        new_col_val=x[col]**dim
        x.insert(0,new_col_name,new_col_val)
        x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=test_size,random_state=0) 
        new_model=LinearRegression()
        new_model.fit(x_train,y_train)
        y_pred=new_model.predict(x_test)
        r2_new=r2_score(y_test,y_pred)
        if r2_new<r2:
            x.drop([new_col_name],axis=1,inplace=True)
        else:
            r2=r2_new
        print(r2_new)
    print(r2)
    
       
```
