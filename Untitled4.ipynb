```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
data=pd.read_csv("EnglandWeather.csv")
df = pd.DataFrame(data)
df
```

```python
format_dates=df['Formatted Date'].str.split(' ')
df.drop(columns=['Formatted Date'],inplace=True)
```

```python
dates=[]
for i in range(len(format_dates)):
    dates.append(format_dates[i][0])
df.insert(column='Date',loc=0,value=dates)    
     
```

```python
df
```

```python
df.sort_values(by='Date',inplace=True)
df
```

```python
df.drop(columns=['Summary',"Precip Type"],inplace=True)
```

```python
df=df.groupby(pd.Grouper(key='Date'),as_index=False).mean()
df
```

```python
plt.scatter(df['Temperature (C)'],df['Humidity']) 
plt.xlabel('Temp')
plt.ylabel('Humidity')
plt.grid()
plt.show()
```

```python
plt.scatter(df['Wind Speed (km/h)'],df['Humidity'])  
plt.xlabel('Wind Speed')
plt.ylabel('Humidity')
plt.grid()
plt.show()
```

```python
df3=df[df['Pressure (millibars)']!=0]
plt.scatter(df3['Pressure (millibars)'],df3['Humidity'])  
plt.xlabel('Pressure (millibars)')
plt.ylabel('Humidity')
plt.grid()
plt.show()
```
